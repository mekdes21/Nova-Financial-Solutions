```python
#necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import talib
import yfinance as yf
import os
```

```python
# Define the path to the raw data directory
data_directory = '../raw_data/yfinance_data/'

# Load the stock data for each company
tsla_data = pd.read_csv(os.path.join(data_directory, "TSLA_historical_data.csv"))
nvidia_data = pd.read_csv(os.path.join(data_directory, "NVDA_historical_data.csv"))
msft_data = pd.read_csv(os.path.join(data_directory, "MSFT_historical_data.csv"))
meta_data = pd.read_csv(os.path.join(data_directory, "META_historical_data.csv"))
goog_data = pd.read_csv(os.path.join(data_directory, "GOOG_historical_data.csv"))
amzn_data = pd.read_csv(os.path.join(data_directory, "AMZN_historical_data.csv"))
aapl_data = pd.read_csv(os.path.join(data_directory, "AAPL_historical_data.csv"))

# Check the column names
print(tsla_data.columns)

```

```python
import yfinance as yf

# Download SPY data for a specific time range
spy_data = yf.download('SPY', start='2020-01-01', end='2024-12-31')

# Print the first few rows of the data to ensure it's downloaded correctly
print(spy_data.head())

# Optional: Check for any issues in the data
if spy_data.empty:
    print("No data was returned for SPY.")
else:
    print("SPY data downloaded successfully.")
```

```python

# Define stock tickers
stock_tickers = ['TSLA', 'NVDA', 'MSFT', 'META', 'GOOG', 'AMZN', 'AAPL']
spy_ticker = 'SPY'

# Dictionary to store stock data
stocks_data = {}

# Download stock data
print("Downloading stock data...")
for ticker in stock_tickers:
    print(f"Downloading data for {ticker}...")
    try:
        data = yf.download(ticker, start='2020-01-01', end='2023-01-01')
        if data.empty:
            print(f"Error: No data for {ticker}. Skipping.")
            continue
        stocks_data[ticker] = data
        print(f"Data for {ticker} downloaded successfully.")
    except Exception as e:
        print(f"Error downloading data for {ticker}: {e}")

# Download SPY data
print("Downloading SPY data...")
try:
    spy_data = yf.download(spy_ticker, start='2020-01-01', end='2023-01-01')
    if spy_data.empty:
        print("Error: No data for SPY.")
except Exception as e:
    print(f"Error downloading SPY data: {e}")

# Apply technical indicators using TA-Lib for each stock
for ticker, df in stocks_data.items():
    print(f"Applying technical indicators for {ticker}...")

    # Ensure there are no NaN values before applying TA-Lib
    close_prices = df['Close'].dropna().values  # Convert to numpy array

    # Ensure it is strictly a 1D array
    close_prices = close_prices.flatten()

    # Moving Average (SMA)
    df['SMA_50'] = talib.SMA(close_prices, timeperiod=50)
    df['SMA_200'] = talib.SMA(close_prices, timeperiod=200)

    # Relative Strength Index (RSI)
    df['RSI'] = talib.RSI(close_prices, timeperiod=14)

    # Moving Average Convergence Divergence (MACD)
    df['MACD'], df['MACD_signal'], df['MACD_hist'] = talib.MACD(close_prices, fastperiod=12, slowperiod=26, signalperiod=9)

# Use PyNance for additional financial metrics (e.g., volatility, returns)
for ticker, df in stocks_data.items():
    print(f"Calculating financial metrics for {ticker}...")
    # Calculate Return on Investment (ROI)
    df['ROI'] = df['Close'].pct_change() * 100

    # Calculate Volatility (Standard deviation of the returns)
    df['Volatility'] = df['ROI'].rolling(window=21).std()  # 21-day rolling window

    # Align stock and market returns by dates (handling missing values)
    stock_returns = df['Close'].pct_change().dropna()

    # Market returns for SPY, aligning dates with the stock data
    market_data = spy_data[['Adj Close']].copy()
    market_data['Market_Return'] = market_data['Adj Close'].pct_change()
    market_returns = market_data['Market_Return'].dropna()

    # Ensure alignment by index (dates)
    aligned_data = pd.concat([stock_returns, market_returns], axis=1, join='inner')
    aligned_data.columns = ['stock_returns', 'market_returns']

    # Calculate Beta using the aligned data
    beta = aligned_data.cov().iloc[0, 1] / aligned_data['market_returns'].var()
    df['Beta'] = beta  # This value will be the same for all dates as it's the overall beta coefficient

# Visualize the Data for all stocks
for ticker, df in stocks_data.items():
    print(f"Visualizing data for {ticker}...")
    plt.figure(figsize=(14,7))

    # Plot Closing Price and SMAs
    plt.subplot(3, 1, 1)
    plt.plot(df.index, df['Close'], label=f'{ticker} Close Price')
    plt.plot(df.index, df['SMA_50'], label='50-day SMA', linestyle='--')
    plt.plot(df.index, df['SMA_200'], label='200-day SMA', linestyle='--')
    plt.title(f'{ticker} Stock Price and Moving Averages')
    plt.legend()

    # Plot RSI
    plt.subplot(3, 1, 2)
    plt.plot(df.index, df['RSI'], label='RSI', color='purple')
    plt.axhline(70, color='red', linestyle='--')  # Overbought level
    plt.axhline(30, color='green', linestyle='--')  # Oversold level
    plt.title(f'{ticker} RSI (Relative Strength Index)')

    # Plot MACD
    plt.subplot(3, 1, 3)
    plt.plot(df.index, df['MACD'], label='MACD', color='blue')
    plt.plot(df.index, df['MACD_signal'], label='MACD Signal', color='orange')
    plt.bar(df.index, df['MACD_hist'], label='MACD Histogram', color='gray', alpha=0.3)
    plt.title(f'{ticker} MACD')

    plt.tight_layout()
    plt.show()

    # Visualize ROI and Volatility
    plt.figure(figsize=(14, 7))

    # Plot ROI (Return on Investment)
    plt.subplot(2, 1, 1)
    plt.plot(df.index, df['ROI'], label=f'{ticker} ROI', color='green')
    plt.title(f'{ticker} ROI (Return on Investment)')

    # Plot Volatility
    plt.subplot(2, 1, 2)
    plt.plot(df.index, df['Volatility'], label=f'{ticker} Volatility', color='red')
    plt.title(f'{ticker} Volatility')

    plt.tight_layout()
    plt.show()

# Save processed data to CSV
data_directory = './processed_data'
if not os.path.exists(data_directory):
    os.makedirs(data_directory)

for ticker, df in stocks_data.items():
    df.to_csv(os.path.join(data_directory, f"{ticker}_processed_data.csv"))
    print(f"Processed data for {ticker} saved.")
```
